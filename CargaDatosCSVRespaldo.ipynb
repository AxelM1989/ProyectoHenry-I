```python
import pandas as pd  # Importar librería pandas
import datetime as dt 
```

```python
# Cargar archivo parquet
producto = pd.read_parquet("producto.parquet")
producto
```

```python
# Ver información del DataFrame. Por ejemplo: Las columnas de categorias solamente tienen 4 datos no nulos cada una de ellas.
producto.info()
```

```python
# Renombrar las columnas. Criterio utilizado: "PascalCase".

producto = producto.rename(columns={"id": "IdProducto", "marca": "Marca", "nombre": "Nombre",
                                    "presentacion": "Presentacion", "categoria1": "Categoria1", "categoria2": "Categoria2", "categoria3": "Categoria3"})
producto
```

```python
# Consultar valores no nulos para saber si se pueden reemplazar. Al realizar este método se observa que no hay algún criterio
#  por el cual se pueda utilizar datos para reemplazar, ya que las categorías son distintas y carecen de datos como para poder unificar

producto.dropna(how="any")
```

```python
# Para complementar al análisis anterior, se realiza una iteración para cada columna de categoría, que nos devuelva los indices(números de filas) de los datos
# no nulos.
def IndiceNoNulo(columna):
    for i, j in enumerate(columna):
        if j == None:
            continue
        else:
            print(i, j)
```

Como conclusion se observa que las únicas dos filas completas son 14903 y 65297.

Las filas 53619 y 55798 tienen datos nulos en Marca, Nombre y Presentación .
Y el resto de las filas tienen datos nulos en Categoria1, Categoria2 y Categoria3.


 Estos datos de falta de categorización en este momento carecen de significado porque, a mi criterio, no se le puede dar un valor significativo en nuestras tablas. Dado que no están completos ni se pueden obtener de otras tablas a través de uniones. No lo desestimaría ya que tal vez es un problema de carga en los datos o que recién se están actualizando y en un futuro si se le podría dar uso y asi cobrarán sentido. Por lo tanto en vez de eliminar las columnas, lo que se va a realizar va a ser un reemplazo en cada valor nulo por la descripción "SinCategoria", para dejar en claro que están sin categorizarse. Este es uno de los criterios que se podrían modificar una vez conversado con los demás integrantes responsables (o con acceso) a los datasets.

A su vez, con respectos a las filas donde los datos nulos son Marca, Nombre y Presentacion, también se reemplaza por SinDatos, ya que se saben las categorías y puede servir para conseguir los valores.

De esta manera, en ninguno de los datos de la tabla producto se perdería ningún Producto (ya que todos tienen IdProducto) y se podría rearmar tal vez la tabla completa.



```python
# Reemplazar columnas de Categorias por "SinCategoria".
producto["Categoria1"].fillna("SinCategoria",inplace=True)
producto["Categoria2"].fillna("SinCategoria",inplace=True)
producto["Categoria3"].fillna("SinCategoria",inplace=True)
```

```python
# Se reemplaza los valores nulos que quedan por "SinDatos".
producto.fillna("SinDatos",inplace=True)
```

```python
# Se comprueba que las filas queden "completadas" con "SinDatos"
print(producto.iloc[53619],producto.iloc[55798])
```

```python
# Ahora se observa que no queda ningún dato nulo en ninguna columna.
producto.info()
```

```python
producto.shape
```

```python
# Se observa si hay duplicados y se eliminan (no hay).
producto.duplicated()
producto.drop_duplicates()
```

```python
# Unificar en la columna "Marca" las primeras letras en mayúsculas de cada palabra.
producto["Marca"]=producto["Marca"].str.title()
producto
```

```python
# Quitar los valores de numeros y unidades/gramos/kilos/milimetros/centimetros cubicos/Litros de la misma columna que el nombre 
# ya que están en la columna de presentación también. Excepto los que hacen alusión a su descripción. Tambien a las promociones
# por ejemplo "Aperitivo Gancia 950 Cc + Aperitivo Gancia Martini Rosso 995 Cc Pa" ó "Estuche Fernet Branca 75 Cc + 2 Vasos".

producto["Nombre"]= producto["Nombre"].astype(str).apply(lambda x:x.strip(" .0123456789UnKgGrmlcLtkx"))
```

```python
# Se reemplaza la descripción de Marca "Sin Marca" por "SinDatos" para unificar criterios y emprolijar.
producto["Marca"].replace("Sin Marca","SinDatos",inplace=True)
```

```python
# Se reasignan orden de las columnas para mayor prolijidad.
producto=producto.reindex(columns=["IdProducto","Nombre","Marca","Presentacion","Categoria1","Categoria2","Categoria3"])
```

```python
# Unificar en la columna "Marca" las primeras letras en mayúsculas de cada palabra. A la columna Provincia la dejamos asi 
# porque no tenemos suficiente información para saber a cual provincia pertenece cada una y no es significativo para nuestro análisis.
# Se lo dejamos como consulta al Data Analyst o quien corresponda sobre si a futuro pretenden realizar algún cambio de criterio. 
sucursal["DescripcionBandera"]=sucursal["DescripcionBandera"].str.title()
sucursal["Localidad"]=sucursal["Localidad"].str.title()
sucursal["NombreSucursal"]=sucursal["NombreSucursal"].str.title()
```
